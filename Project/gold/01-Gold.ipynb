```python
class Gold():
    def __init__(self):
        self.goldSchemaName = "gold"
        self.goldViewName = 'high_priority_customers_vw'
    
        self.viewQuery = """
            WITH full_joined AS (
                SELECT s.CUST_ID, 
                concat(c.CUST_FIRST_NAME, ' ', c.CUST_LAST_NAME) AS CUST_NAME, 
                c.CUST_MAIN_PHONE_NUMBER, 
                c.CUST_EMAIL, 
                s.TIME_ID, 
                t.CALENDAR_YEAR, 
                s.AMOUNT_SOLD 
                FROM `selectiva-project`.edw.sales_edw AS s
                INNER JOIN `selectiva-project`.edw.times_edw AS t
                USING (TIME_ID)
                INNER JOIN (SELECT * FROM `selectiva-project`.edw.customers_edw WHERE CUST_VALID = 'A') AS c
                USING(CUST_ID)
            )

            SELECT 
                CUST_NAME, 
                CUST_MAIN_PHONE_NUMBER, 
                CUST_EMAIL,  
                SUM(AMOUNT_SOLD) AS TOTAL_SOLD
            FROM full_joined
            GROUP BY ALL
            HAVING TOTAL_SOLD > 1500
            """
    def createView(self):
        spark.sql(f""" 
                  CREATE OR REPLACE VIEW `selectiva-project`.{self.goldSchemaName}.{self.goldViewName} AS ({self.viewQuery})
                """) 
```

```python
v = Gold()
v.createView()
```
